```
#python Matplotlib Project
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

#Reading IPL Data
matches_raw=pd.read_csv('matches.csv')
print("Different columns of IPL Data:")
print(matches_raw.info())

plt.rcParams.update({'font.size': 10})
#Showing no of Matches per season
matches_per_season = matches_raw.groupby('season').id.count()
plt.figure(figsize=(10,5))
plt.bar(matches_per_season.index,matches_per_season,color="g",)
plt.title("Matches Per Season")
plt.xlabel('Year')
plt.ylabel('Matches')
plt.show()

#Showing Highest Toss Winners
#print(matches_raw['toss_winner'].value_counts())
w=[0.7,0.6,0.4,0.3,0.2,]
plt.figure(figsize=(12,6))
plt.bar(list(matches_raw['toss_winner'].value_counts()[0:5].keys()),list(matches_raw['toss_winner'].value_counts()[0:5]),width=w,color=["blue","purple","yellow","red","red"])
plt.title("Most No of Toss winners")
#plt.set(xlabel="Player",ylabel="Awards")
plt.xlabel('Team')
plt.ylabel('No of Toss wins')
plt.show()


#Showing Most Man of Match Awards
#Showing Top 5 Players
#print(matches_raw['player_of_match'].value_counts())
plt.figure(figsize=(10,5))
plt.bar(list(matches_raw['player_of_match'].value_counts()[0:5].keys()),list(matches_raw['player_of_match'].value_counts()[0:5]),width=w,color=["red","red","blue","orange","yellow"])
plt.title("Most Man of the Match Awards")
plt.xlabel('Players')
plt.ylabel('Man of the Matche Awards')
plt.show()
#print(matches_raw['player_of_match'].value_counts())

plt.rcParams.update({'font.size': 8})
#No of Teams who won batting First
batting_first=matches_raw[matches_raw['win_by_runs']!=0]
#print(batting_first['winner'].value_counts())
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title("Batting First")
plt.show()

#No of Teams who won Bowling First
bowling_first=matches_raw[matches_raw['win_by_wickets']!=0]
#print(bowling_first['winner'].value_counts())
plt.figure(figsize=(7,7))
plt.pie(list(bowling_first['winner'].value_counts()),labels=list(bowling_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title("Bowling First")
plt.show()

#Most Wins by a team
most_wins=matches_raw['winner'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(list(matches_raw['winner'].value_counts()),labels=list(matches_raw['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title("Most Wins")
plt.show()
```
